```python
#10 Sort
'''
Sort array_up by first column
Sort array_down by first column desc
'''

sorted_index_1stcol_arr_up = array_up[:, 0].argsort() #sort arr_up ascending order
sorted_index_1stcol_arr_down_desc = array_down[:, 0].argsort()[::-1] #sort arr_down descending order

array_up_sorted = array_up[sorted_index_1stcol_arr_up]
array_down_sorted_desc = array_down[sorted_index_1stcol_arr_down_desc]
print(array_up_sorted[:20])
print(array_down_sorted_desc[:20])
```

```python
#11. Concatenate arrays

array_joint = np.concatenate([array_up_sorted, array_down_sorted_desc], axis = 0)

print(array_joint[:20])  #first 20 rows
print(array_joint[-20:]) #last 20 rows
```

```python
#13 Function
'''
Transform input array:
      Xi*2 if Xi<Xavg
      Xi/4 if Xi<=Xavg

Input parameter - array
'''

def transformer_arr(array, column_index):
  average_arr = np.average(array[:, column_index])
  result = np.copy(array)  #copy in order to process values once

  result[array[:, column_index] < average_arr, column_index] *= 2
  result[array[:, column_index] >= average_arr, column_index] /= 4

  return result
```

```python
#14 Applying function to col2
print(array_joint[:20])

array_final_hw_l2 = transformer_arr(array = array_joint, column_index = 2)

print('Transformed dataset: \n',  array_final_hw_l2[:20])
```
