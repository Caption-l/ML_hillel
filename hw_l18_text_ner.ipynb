```python
#Встановлення бібліотек
import spacy
from spacy import displacy
```

Спочатку тестування на англ тексті

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
text = "Apple is looking at buying U.K. startup for $1 billion"

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
displacy.render(doc, style="ent")
```

Переходимо до української лінгвістичної моделі

```python
#Завантаження моделі spaCy для обробки текстів українською мовою; модель натренована на текстах з новин
nlp = spacy.load("uk_core_news_sm")

#Беремо довільний текст з новин технологій
text = "Найбільші технологічні гіганти, як-от Google, IBM та Intel, вкладають значні кошти у квантові дослідження, які змінять мобільну техніку. Квантові комп'ютери мають потенціал для експоненціального збільшення обчислювальної потужності та ефективності, що важливо для смартфонів. Про це пише The Smartphone Magazine. У світі технологій, що постійно розвивається, квантові комп'ютери стають майбутніми гравцями, які змінять правила гри. У той час як традиційні комп'ютери покладаються на двійкові біти, квантові комп'ютери використовують міць квантових бітів, або кубітів, які можуть існувати в декількох станах одночасно."

#Обробка тексту за допомогою nlp-моделі
doc = nlp(text)

displacy.render(doc, style="ent")

#Виводимо інформацію про span-об'єкти, такі як початок і кінець об'єкту у рядку, мітка об'єкту
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)

```

```python
#Для детальнішого аналізу (по токенах) використаємо внутрішній цикл
for ent in doc.ents:
    print(f"Сутність: {ent.text} ({ent.label_})")
    for token in ent:
        print(f"  {token.text} {token.pos_} {token.tag_} {token.dep_} {token.is_alpha} {token.is_stop}")
```

У даному випадку модель виділила тільки латиномовні абревіатури, що з одного боку дещо недостатньо, а з іншого - позитивний момент, оскільки враховує те, що серед слів українською мовою можуть бути і терміни на латинці.

Як бачимо, аналізу на частини мови при тренуванні не проводилося.
