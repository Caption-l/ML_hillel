**4-5. Які средні значення та средньоквадратичні відхилення віку тих, хто отримує більше 50K на рік (ознака *salary*) і тих, хто отримує менше 50K на рік?**

**6. Чи правда, що люди, які отримують більше 50k, мають як мінімум вищу освіту? (ознака *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
high_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']

(data[data['salary'] == '>50K']['education'].value_counts()[high_education]).sum() == data[data['salary'] == '>50K'].shape[0]
```

**8. Серед кого більша доля багатіших (>50K): серед одружених чи не одружених чоловіків (ознака *marital-status*)? За одружених вважаємо тих, у кого *marital-status* починається із *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), всі інші - не одружені.**

```python
married = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']

#Create new column with shortened marital-status
data['shortened-marital-status'] = data['marital-status'].apply(lambda x: 'married' if x in married else 'not_married')

#Group by shortened marital status
print(data[data['sex'] == 'Male'].groupby('shortened-marital-status')['salary'].value_counts(normalize=True), "\n")
#print(data[data['sex'] == 'Male'].groupby('marital-status')['salary'].value_counts(normalize=True))

#Find ratio rich men among married men
rich_married_ratio = data[data['sex'] == 'Male'].groupby('shortened-marital-status')['salary'].value_counts(normalize=True).\
                                                  loc['married', '>50K']

#Find ratio rich men among unmarried men
rich_not_married_ratio = data[data['sex'] == 'Male'].groupby('shortened-marital-status')['salary'].value_counts(normalize=True).\
                                                  loc['not_married', '>50K']

#Compare ratios
if rich_married_ratio > rich_not_married_ratio:
  print("Доля багатих більша серед одружених чоловіків і становить", round(rich_married_ratio,2), "проти",
        round(rich_not_married_ratio,2), "серед неодружених чоловіків")
elif rich_married_ratio < rich_not_married_ratio:
  print("Доля багатих більша серед неодружених чоловіків і становить", round(rich_not_married_ratio,2), "проти",
        round(rich_married_ratio,2), "серед одружених чоловіків")
else:
  print("Долі багатих однакові серед одружених і неодружених чоловіків і становлять", rich_married_ratio)
```

**9. Яку максимальну кількість годин в тиждень працює людина (ознака *hours-per-week*)? Скільки людей працюють таку кількість годин і який відсоток серед них багатих?**

```python
max_work_hours_per_week = data['hours-per-week'].max()

print('Максимальна кількість годин в тиждень:', max_work_hours_per_week)

print('Кількість людей, які працюють максимальну кількість годин:', (data['hours-per-week'] == max_work_hours_per_week).sum())

print('Серед них багатих є', round(100 * data[data['hours-per-week'] == max_work_hours_per_week]['salary'].
                                   value_counts(normalize = True)['>50K'], 2), "%")
```
